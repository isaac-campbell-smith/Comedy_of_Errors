```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from src.user_recommender import ReviewRecommender
from src.item_reccomender import ItemRecommender
#from src.make_ratings_matrix import ratings_mat
```

```python
specials = pd.read_pickle('data/working_corpus.pkl')
reviews = pd.read_pickle('data/working_reviews.pkl')
```

```python
tf = TfidfVectorizer(max_features=10000, max_df=0.9, ngram_range=(3,10))
tf_vector = tf.fit_transform(specials.lem_tokens.apply(lambda ls: ' '.join(ls))).toarray()
```

```python
user_recommender = ReviewRecommender()
user_recommender.fit(reviews, specials.special, specials.genre, tf_vector)
```
